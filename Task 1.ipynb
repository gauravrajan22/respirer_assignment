#  Meter readings analysis for seasonal, monthly, daily, and hourly measurements

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading dataset
data = pd.read_excel('cons_data_all_res.xlsx')
```

```python
print(data.head())
```

```python
for col in range(1,13):
    fig=plt.figure(figsize=(12, 8))

    # Format column names
    col_name = f'r{col}'
    plt.title("Meter Reading ({col_name})")
    # Seasonal plot
    plt.subplot(2, 2, 1)
    data.groupby('season')[col_name].mean().plot(kind='bar')
    plt.title('Seasonal Meter Readings')
    plt.xlabel("")

    # Monthly plot
    plt.subplot(2, 2, 2)
    data.groupby('moy_date')[col_name].mean().plot()
    plt.title('Monthly Meter Readings')
    plt.xlabel("Month")

    # Daily plot
    plt.subplot(2, 2, 3)
    data.groupby('doy_date')[col_name].mean().plot()
    plt.title('Daily Meter Readings')
    plt.xlabel("Day")

    # Hourly plot
    plt.subplot(2, 2, 4)
    data.groupby('hour')[col_name].mean().plot()
    plt.title('Hourly Meter Readings')
    plt.xlabel("Hour")

    plt.tight_layout()
    plt.subplots_adjust(top=0.9)
    fig.suptitle(f'Meter Readings Analysis - {col_name}', fontsize=16)
    plt.show()
```
